```python
import numpy as np
a=[[1,2,3],[4,5,6],[7,8,9]]
b=[[1,3,4],[4,2,9],[1,2,3]]
c=np.array(a)
d=np.array(b)
print(c+d)
#multiplication
e=np.matmul(a,b)
print(e)
f=np.power(a,b)
print(f)
g=np.divide(a,b)
print(g)
```
